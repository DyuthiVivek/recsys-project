```python
cols_to_keep = ['user_id', 'product_id', 'event_time']
df = df[cols_to_keep] 

df['user_id'] = df['user_id'].astype('category')
df['event_time'] = pd.to_datetime(df['event_time'], format='%Y-%m-%d %H:%M:%S UTC')
df.sort_values(by=['user_id', 'event_time'], inplace=True, ignore_index=True)
```

```python
df['time_str'] = df['event_time'].dt.strftime('%Y-%m-%d %H:%M:%S')
```

```python
all_products = df['product_id'].dropna().unique().tolist()
```

```python
def sample_negatives(target, all_prods, k):
    negs = set()
    while len(negs) < k:
        candidate = str(random.choice(all_prods))
        if candidate != target:
            negs.add(candidate)
    return list(negs)
```

```python
import json
import random

window_size = 15
num_negatives = 100


df['product_id'] = df['product_id'].astype(str)
df['time_str'] = df['time_str'].astype(str) 

output_filepath = 'e-commerce_sequences.jsonl'

# 3 GB limit in bytes (3 * 1024 * 1024 * 1024)
MAX_BYTES = 3 * 1024 * 1024 * 1024 
bytes_written = 0
records_saved = 0
limit_reached = False

with open(output_filepath, 'w', encoding='utf-8') as f:
    
    # Iterate over the groups
    for user, group in df.groupby('user_id', observed=True):
        if limit_reached:
            break
            
        if len(group) < (window_size + 2):
            continue
            
        products = group['product_id'].tolist()
        timestamps = group['time_str'].tolist()
        
        start_idx = window_size + 1 
        
        for i in range(start_idx, len(products)):
            target_item = products[i]
            
            negative_items = sample_negatives(target_item, all_products, num_negatives)
            
            record = {
                'user_id': str(user),
                'slow_memory_ids': products[0 : i - window_size],
                'slow_id_timestamps': timestamps[0 : i - window_size],
                'fast_memory_ids': products[i - window_size : i],
                'fast_id_timestamps': timestamps[i - window_size : i],
                'target_item': target_item,
                'target_timestamp': timestamps[i],
                'negative_items': negative_items
            }
            
            line_to_write = json.dumps(record) + '\n'
            
            line_size = len(line_to_write.encode('utf-8'))
            
            # Check if writing this line exceeds our 3 GB limit
            if bytes_written + line_size > MAX_BYTES:
                limit_reached = True
                break  # Break inner loop
            
            f.write(line_to_write)
            bytes_written += line_size
            records_saved += 1

print(f"Done! Dataset saved memory-safely to '{output_filepath}'.")
print(f"Total disk space used: {bytes_written / (1024**3):.4f} GB")
print(f"Total data points saved: {records_saved:,}")
```
